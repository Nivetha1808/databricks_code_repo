```python
# Create widget
dbutils.widgets.text("param1", "default_value", "Your input parameter")

# Read value
param_value = dbutils.widgets.get("param1")
print("Widget value received:", param_value)

# run a Spark operation using widget
try:
    df = spark.sql(f"SELECT * FROM default.{param_value} LIMIT 5")
    df.show()
    result = f"Child notebook: Loaded table {param_value}"
except Exception as e:
    result = f"Child notebook error: {str(e)}"

# Return result to parent
dbutils.notebook.exit(result)
```
